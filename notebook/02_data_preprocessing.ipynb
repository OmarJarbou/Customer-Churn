# Stage 2 — Data Preprocessing

This stage focuses on cleaning and preparing the dataset for machine learning models.

The preprocessing steps include:
- handling missing values
- fixing data inconsistencies
- converting categorical variables into numerical format
- encoding features
- preparing the dataset for training and testing

```python
# =====================================
# Import Libraries
# =====================================
import pandas as pd
import numpy as np

df = pd.read_csv("../data/WA_Fn-UseC_-Telco-Customer-Churn.csv")

df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

```python

# Check for empty strings
(df == " ").sum()
```

```python
# TotalCharges should be numerical, but it is stored as text because of empty values.
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")

# Check for missing values again
df.isnull().sum()
```

```python
# Check rows where TotalCharges became missing
df[df["TotalCharges"].isnull()]
```

```python
# Replace missing TotalCharges with 0:
df["TotalCharges"] = df["TotalCharges"].fillna(0)

# Check for missing values again
df.isnull().sum()
```

```python
# Check for duplicates
df.duplicated().sum()
```

```python
df = df.drop_duplicates()

# Check again
df.duplicated().sum()
```

```python
# customerID is only an identifier. It does not help the model predict churn.
df = df.drop("customerID", axis=1)

# Check duplicate feature rows after removing customerID
print("Duplicate feature rows after removing customerID:", df.duplicated().sum())
```

```python
# Check unique values in categorical columns
categorical_cols = df.select_dtypes(include=["object", "string"]).columns

for col in categorical_cols:
    print(col)
    print(df[col].unique())
    print("-" * 40)
```

```python
# Convert target variable Churn to binary
df["Churn"] = df["Churn"].map({"Yes": 1, "No": 0})

# Check
df["Churn"].value_counts()
```

```python
# Convert binary categorical variables to binary data system
binary_cols = [
    "gender",
    "Partner",
    "Dependents",
    "PhoneService",
    "PaperlessBilling"
]

df["gender"] = df["gender"].map({"Male": 1, "Female": 0})

for col in ["Partner", "Dependents", "PhoneService", "PaperlessBilling"]:
    df[col] = df[col].map({"Yes": 1, "No": 0})

# Check results
df[binary_cols].head()
```

```python
# Apply One-Hot Encoding to multi-category variables
multi_category_cols = [
    "MultipleLines",
    "InternetService",
    "OnlineSecurity",
    "OnlineBackup",
    "DeviceProtection",
    "TechSupport",
    "StreamingTV",
    "StreamingMovies",
    "Contract",
    "PaymentMethod"
]

df = pd.get_dummies(df, columns=multi_category_cols, drop_first=True)
```

```python
# Check the new shape
df.shape
```

```python
# Check datatypes
df.dtypes
```

```python
# Convert only boolean columns to integers
bool_cols = df.select_dtypes(include=["bool"]).columns
df[bool_cols] = df[bool_cols].astype(int)

# Check datatypes
df.dtypes
```

```python
df.head()
```

```python
df.info()
```

```python
print("Missing values:", df.isnull().sum().sum())
print("Shape:", df.shape)
print("Duplicate feature rows:", df.duplicated().sum())
# Important point: after dropping customerID, two different customers may look identical 
# because all their features are the same. So these are not necessarily “duplicate records” 
# in the original dataset — they may be different customers with the same profile.
```

```python
df.to_csv("../data/telco_churn_cleaned_encoded.csv", index=False)
```
